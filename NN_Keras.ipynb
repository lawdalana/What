```python
from keras.models import Sequential
from keras.layers import Dense, Activation, Flatten
from keras.layers.normalization import BatchNormalization
from keras.utils import np_utils
import numpy as np
```

```python
Trainfile = "C:/Data/NN_Train.csv"
Testfile = "C:/Data/NN_Test.csv"
```

```python
### import data #####
novelid_train = list()
label_train = list()
data_train = list()
with open(Trainfile,"r+") as test:
    for i in test:
        i = i.replace("\n","")
        buf = i.split(",")
        novelid_train.append(buf[0])
        label_train.append(buf[-1])
        data_train.append(buf[1:5])

novelid_test = list()
label_test = list()
data_test = list()
with open(Testfile,"r+") as test:
    for i in test:
        i = i.replace("\n","")
        buf = i.split(",")
        novelid_test.append(buf[0])
        label_test.append(buf[-1])
        data_test.append(buf[1:5])
```

```python
data_train = np.array(data_train)
data_train = data_train.reshape(data_train.shape[0], 4)
data_test  = np.array(data_test)
data_test = data_test.reshape(data_test.shape[0], 4)
label_train = np.array(label_train)
label_test = np.array(label_test)
# label_train = np_utils.to_categorical(label_train, 2)
# label_test = np_utils.to_categorical(label_test, 2)
```

```python
### Layer #####
model = Sequential()
model.add(Dense(10, input_dim=4, activation='relu'))
# model.add(BatchNormalization())
model.add(Dense(5, activation='relu'))
# model.add(BatchNormalization())
model.add(Dense(1, activation='relu'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(data_train, label_train, epochs=100, batch_size=64)
```

```python
# print model.summary()
# score = model.evaluate(data_test, label_test, verbose=0)
# print(‘Test score:’, score[0])
# print(‘Test accuracy:’, score[1])
scores = model.evaluate(data_test, label_test)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
